# 3. Country Analysis

```python
# Calculating dataset where  funding_round_type='venture' and Status!='Closed' and raised_amount_used
# between 5 to 15 million USD per  investment round
top9_dataset = master_frame[(master_frame['funding_round_type']=='venture') & (master_frame['status']!='closed') & (master_frame['raised_amount_usd']>5000000) & (master_frame['raised_amount_usd']<15000000)]
top9_dataset
```

```python
# top9_dataset_countrywise denoted the group by data country wise where fund is sorted in descending order to get top investment s countries
top9_dataset_countrywise = top9_dataset.groupby(by=['country_code'])['raised_amount_usd'].sum().sort_values(ascending=False).to_frame('sum').reset_index(drop=False)
top9_dataset_countrywise 
```

```python
#Logic for WebScraping the given Link to get the Country_List where English is an official Language and store ina file 
# named as english_speaking_country.csv

import pandas as pd
import requests
from bs4 import BeautifulSoup


r=requests.get('https://en.wikipedia.org/wiki/List_of_territorial_entities_where_English_is_an_official_language')
soup = BeautifulSoup(r.content,features='html.parser')
soup.prettify()

data=[]
english_speaking_country=[]
tables = soup.findAll('table',attrs={'class':'wikitable'})
for table in tables:
    table_body = table.find('tbody')
    rows = table_body.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        if len(cols)==6 and cols is not None and cols[2] is not None:
            english_speaking_country.append(cols[2])

print('English Speaking Country: {}'.format(english_speaking_country),end='\n')
english_country_list = pd.DataFrame({'country_code':english_speaking_country})
english_country_list.to_csv('english_speaking_country.csv',index=False)
```

```python
# Calculating dataset where  funding_round_type='venture' and Status ='Closed' and raised_amount_used
# between 5 to 15 million USD per  investment round and countries are USA,GBR ,CAN
top3_invest_country_dataset = top9_dataset[(top9_dataset['country_code'].isin (top9_dataset_countrywise_value['country_code'][0:3]))]
top3_invest_country_dataset


```

```python
#Find the value of primary_sector from the top9_dataset based on category_List
top3_invest_country_dataset['primary_sector'] = top3_invest_country_dataset['category_list'].fillna('').apply(lambda x:x.split('|')[0])
top3_invest_country_dataset
```

```python
# Find the main_sector for earch primary sector in the top3_country_dataset
def test(row):
    dd = main_sector[main_sector['category_list']==row]
    if dd.shape[0]>0:
        return (dd == 1).idxmax(axis=1).values[0] 
    else:
        return None
    
top3_invest_country_dataset.loc[:,'main_sector']= top3_invest_country_dataset['primary_sector'].apply(test)
```

# 5. Plots

```python
# A plot showing the fraction of total investments (globally) in angel, 
# venture, seed, and private equity, and the average amount of investment 
# in each funding type. This chart should make it clear that a certain funding 
# type (FT) is best suited for Spark Funds.
%matplotlib inline
fraction_investment_type=master_frame.funding_round_type.value_counts().to_frame('no of funding').reset_index(drop=False)
fraction_investment_type['fraction'] =  fraction_investment_type['no of funding'].apply(lambda x:x/fraction_investment_type['no of funding'].sum())
fraction_investment_type['avg_amount_invested'] = fraction_investment_type['index'].apply(lambda x:master_frame[master_frame['funding_round_type']==x]['raised_amount_usd'].mean())
fraction_investment_type['avg_amount_invested'] = fraction_investment_type['avg_amount_invested'].apply(lambda x:round(x))
fraction_selected_investment_type=fraction_investment_type[(fraction_investment_type['index']=='venture') | (fraction_investment_type['index']=='seed') | (fraction_investment_type['index']=='angel') | (fraction_investment_type['index']=='private_equity')]
fraction_selected_investment_type = fraction_selected_investment_type.rename(columns={'index':'investment_type'})
```

```python
fraction_selected_investment_type.plot(y='fraction',x='investment_type',kind='bar', color='brown')
fraction_selected_investment_type
```

```python
# A plot showing the fraction of total investments (globally) in angel, 
# venture, seed, and private equity, and the average amount of investment 
# in each funding type. This chart should make it clear that a certain funding 
# type (FT) is best suited for Spark Funds.

import os
import numpy as np
import matplotlib.pyplot as plt

investment_type = fraction_selected_investment_type['investment_type'].tolist()
fraction = fraction_selected_investment_type['fraction'].tolist()
avg_amount_invested = fraction_selected_investment_type['avg_amount_invested'].tolist()

fig, ax = plt.subplots()
plt.ylabel('Investment Type')  
width = 0.75 # the width of the bars 
ind = np.arange(len(investment_type))  # the x locations for the groups
ax.barh(ind, fraction, width, color="brown")
for i, v in enumerate(avg_amount_invested):
    ax.text(.55, i + .05, str(v), color='green', fontweight='bold')
ax.set_yticks(ind+width/2)
ax.set_yticklabels(investment_type, minor=False)
plt.title('Fraction of total investments Based on Fund Type')
plt.xlabel('Fraction')
plt.show()
```
